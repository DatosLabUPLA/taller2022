importando librerias

```python
import requests
import csv
from bs4 import BeautifulSoup as bs
import pandas as pd
```

*-----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------*

inicializando Scimago journal

```python
url='https://www.scimagojr.com/journalrank.php'
scimago=requests.get(url)
```

```python
print(scimago.content)
```

```python
import bs4 as b
soup = bs(scimago.text, 'lxml')
```

```python
soup.find('div', attrs={'class':'ranking_body'})
```

```python
extract=soup.find_all('td', class_='tit')
```

```python
extract
```

```python
journal_category=list()

for i in extract:
  journal_category.append(i.text)
```

```python
print(journal_category)
```

```python
extract_orde=soup.find_all('td', class_='orde')
```

```python
print(extract_orde)
```

```python
document=list()
for i in extract_orde:
    document.append(i.text)
```

```python
print(document)
```

```python
testing=list()
test=soup.find_all('td')
for i in test:
  testing.append(i.text)

```

```python
testing
```

*creacion dataframe y csv*

```python
data=pd.DataFrame({'Categoria Journal':journal_category ,'Numero de documentos':document } )
```

```python
data
```

```python
data.to_csv('DATA/journal.csv', index=False, encoding='utf-8')
```
