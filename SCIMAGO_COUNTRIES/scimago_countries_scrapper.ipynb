llamo a las librerias

```python
import requests
import csv
from bs4 import BeautifulSoup as bs
import pandas as pd
```

inicializo la pagina

```python
url="https://www.scimagojr.com/countryrank.php"
scimago=requests.get(url)
```

```python
print(scimago.content)
```

```python
import bs4 as b
sopa=bs(scimago.text, 'lxml')
```

```python
sopa.find('div', attrs={'class':'ranking_body'})
```

ocupo un find_all para extraer todos los paises registrados y el numero de documentos que tienen cada uno

```python
cazuela=sopa.find_all('td', class_='tit')
cazuela
```

```python
paises=list()

for i in cazuela:
  paises.append(i.text)

print(paises)
```

```python
ramen=sopa.find_all('div', class_='dentro_td')
print(ramen)
```

```python
ndocu=list()
for i in ramen:
  ndocu.append(i.text)

print(ndocu)
```

```python
minestrone=sopa.find_all('td', class_='cel_orde')
print(minestrone)
```

aqui converti todas las etiquetas TD de la pagina en una lista y las almacene el la variable prueba

```python
prueba=list()
test=sopa.find_all('td')
for i in test:
  prueba.append(i.text)

prueba
```

lo siguiente fue verificar en que posicion empezar, en este caso la 4ta contando desde 0 y que vaya de 9 en 9, ya que es la distancia entre las variables del mismo tipo

```python
citables=[]

for i in range(4,2156,9):
  citables.append(prueba[i])
citables
```

```python
citaciones=[]

for i in range(5,2157,9):
  citaciones.append(prueba[i])
citaciones
```

```python
auto=[]

for i in range(6,2158,9):
  auto.append(prueba[i])
auto
```

```python
citapd=[]

for i in range(7,2159,9):
  citapd.append(prueba[i])
citapd
```

```python
index=[]

for i in range(8,2160,9):
  index.append(prueba[i])
index
```

una vez toda la tabla trabajada se crea el dataframe, contando desde 1 hasta 241

```python
data=pd.DataFrame({'Pais':paises,'Numero de documentos':ndocu, 'Documentos Citables':citables, 'Citaciones':citaciones, 'Autocitas':auto, 'Citas por Documento':citapd, 'H index':index}, index=list(range(1,241)))
data
```

```python
data.to_csv("DATA/listado_paises.csv")
```
